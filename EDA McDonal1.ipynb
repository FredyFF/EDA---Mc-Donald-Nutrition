# Exploratory Data Analysis (EDA) Mc Donald Nutrition Menu

This analysis aims to explore the nutritional data of well-known fast food companies. This data raises several questions regarding the types of customers who come. And of course, carrying out this analysis will make it easier for readers to know the methods used to gain insight

This dataset provides a nutrition analysis of every menu item on the US McDonald's menu, including breakfast, beef burgers, chicken and fish sandwiches, fries, salads, soda, coffee and tea, milkshakes, and desserts.

## import Dataset

```python
import pandas as pd
df_mcd = pd.read_csv("menu.csv")

# get info dataset
df_mcd.info()
```

## In each dish, what menu is suitable for customers who are on a diet ?

When dieting, we should avoid the following foods:
- High-fat and fried foods: Foods such as French fries, fried chicken and other fried foods are high in calories
  and saturated fat. 
- Sweet foods and drinks: Foods and drinks that contain high levels of sugar such as candy, ice cream, cakes,
  biscuits, fizzy drinks and sweet tea should be avoided. 
- Processed foods: Processed foods are generally rich in salt, sugar and trans fats. 
- Alcoholic drinks: Alcohol can hinder the weight loss process because it is rich in calories. 
- White flour-based products: White flour, found in white bread, pasta, and cakes, has been through a refining
  process that removes fiber and nutrients. 
- Foods containing coconut milk: Foods that use coconut milk usually have high levels of fat and calories.

so we can see foods with low standards from foods that contain these things.
